### Inheritance

Objects are often very similar in that they share a common logic, but they're not entirely the same. We can reuse the common logic and extract the unique logic into a separate class using __Inheritance__. 

Inheritance allows an instance to inherit attributes and methods from a parent class. This is useful because it allows us to create (child) subclasses. We get all the functionality of the parent class plus the ability to overwrite or add new functionality without affecting the parent class in any way. 

For example, if we want to specify different types of Greek gods, i.e. Olympians and Titans. To create a subclass, simply define a new class and pass the parent class as the argument.

```python
# Create the class
class God:
    
    raise_amount = 1.05
    
    def __init__(self, name, domain, symbol, tribute):
        self.name = name
        self.domain = domain
        self.symbol = symbol
        self.tribute = tribute
        
    def tribute_raise(self):
        self.tribute = int(self.tribute * self.raise_amount)


class Olympian(God):
    pass

class Titan(God):
    pass

# Create instances of the class
oly_1 = Olympian('Zeus', 'Sky', 'Lightning Bolt', 100000)
titan_1 = Titan('Kronos', 'Time', 'Sickle', 30000)
```

### Polymorphism with Inheritance

__Polymorphism__ lets us define methods in the child class that have the same name as the methods in the parent class. We might have a parent class and a few child classes that inherit from it. In inheritance, the child class inherits the methods from the parent class. However, we may need to modify a method that has been inherited in a child class (for whatever reason, this method doesn't fulfill the need of the child class or if there are multiple child classes and only a few require this redefinition without affecting the parent class code). Using method overriding, polymorphism allows us to defines methods in the child class that have the same name as the methods in the parent class. 

In polymorphism, Python first checks the object’s class type and executes the appropriate method when we call the method. To override a method in the base class, a sub class needs to define a method of same signature. (i.e same method name and same number of parameters as method in base class).

```python
class God():
    
    def __init__(self):
        self.__x = 1

    def print_statement(self):
        print("God")


class Titan(God):

    def __init__(self):
        self.__y = 1

    def print_statement(self):
        print("Titan")

test = Titan()
test.print_statement()
```

Here we are overriding the print_statement() method from the God base class. If we comment out the print_statement() method in the Titan class, now the print_statement() method from the God base class will run.

### Polymorphism with Class Methods

Polymorphism with class methods is useful when we want objects to perform the same action in different ways. In the example below, we have two classes: Circle and Rectangle. Calculating their area would be a function they both share, but they require different formulas. So, we can create a calculate_area() instance method under both classes and a function outside these that takes an object and calls that object's calculate_area() method to implement polymorphism. 

```python
class Circle:
    pi = 3.14

    def __init__(self, redius):
        self.radius = redius

    def calculate_area(self):
        print("Area of circle :", self.pi * self.radius * self.radius)

class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def calculate_area(self):
        print("Area of Rectangle :", self.length * self.width)

# function
def area(shape):
    # call action
    shape.calculate_area()

# create object
cir = Circle(5)
rect = Rectangle(10, 5)

# call common function
area(cir)
area(rect)
```

### Encapsulation

It describes the idea of wrapping data and the methods that work on data within one unit. For example, a class is an example of encapsulation as it encapsulates all its methods and variables into a single unit. __Encapsulation__ is achieved when each object keeps its internal representation hidden from outside, i.e. private. Other objects don't have direct access to this state. Instead, they can only call a list of public functions - called __methods__. 

Encapsulation acts as a protective layer by hiding data from the outside world. We can restrict access to methods and variables from outside, and it can prevent the data from being modified by accidental or unauthorized modification. 

The God class we created earlier implements encapsulation. We can make a variable private by prefixing a double underscore before a variable:

```python
class God:
    
    raise_amount = 1.05
    
    def __init__(self, name, domain, symbol):
        self.name = name
        self.domain = domain
        self.__symbol = symbol
        
oly_1 = God('Zeus', 'Sky', 'Lightning Bolt')
```

```python
print(oly_1.name)
print(oly_1.domain)
print(oly_1.__symbol)
```

As shown by the 'AttributeError', the name and domain variables are public and accessible, but symbol is a private variable, so it cannot be accessed from outside the class. Trying to access it will raise an error.

### Abstraction

__Abstraction__ is the concept whererin the user is kept unaware of the basic implementation of a function property. The user is only able to view basic functionalities whereas the internal details are hidden. It is an extension of encapsulation. 

For example, people do not think of a car as a set of thousands of individual parts. Instead they see it as a well-defined object with its own unique behavior. This abstraction allows people to use a car to drive without knowing the complexity of the parts that form the car. They can ignore the details of how the engine transmission, and braking systems work. Instead, they are free to utilize the object as a whole.

A powerful way to manage abstraction is through the use of hierarchical classification. This allows us to layer the semantics of complex systems, breaking them into more manageable pieces. From the outside, a car is a single object. Once inside, you see that the car consists of several subsystems: steering, brakes, sound system, seat belts, etc. In turn, each of these subsystems is made up of smaller units.

The point is that we manage the complexity of the car (or any other complex system) through the use of hierarchical abstractions.

In Python, abstraction can be achieved by having/using abstract classes and methods in our programs. An abstract method is a method that is declared, but does not contain implementation. An abstract method in a base class identifies the functionality that should be implemented by all its subclasses. However, since the implementation of an abstract method would differ from one subclass to another, often the method body comprises just a pass statement. Every subclass of the base class will ride this method with its implementation. A class containing abstract methods is called abstract class. Python provides the abc module to use the abstraction in the Python program, syntax as:

```python
from abc import ABC, abstractmethod

class Abs_class(ABC):
    
    #normal method
    def method(self):
        #method definition
        print('Statement')
        
    @abstractmethod
    def Abs_method(self):
        #Abs_method definition
        pass
```

```python
from abc import ABC, abstractmethod

class Absclass(ABC):

    def print(self,x):
        print("Passed value: ", x)
        
    @abstractmethod
    def task(self):
        print("We are inside Absclass task")

class test_class(Absclass):
    def task(self):
        print("We are inside test_class task")

class example_class(Absclass):
    def task(self):
        print("We are inside example_class task")

#object of test_class created
test_obj = test_class()
test_obj.task()
test_obj.print(100)
 
#object of example_class created
example_obj = example_class()
example_obj.task()
example_obj.print(200)
 
print("test_obj is instance of Absclass? ", isinstance(test_obj, Absclass))
print("example_obj is instance of Absclass? ", isinstance(example_obj, Absclass))
```

Absclass is the abstract class that inherits from the ABC class from the abc module. It contains an abstractmethod task() and a print() method which are visible by the user. Two other classes inheriting from this abstract class are test_class and example_class. Both of them have their own task() method (extension of the abstract method).

After the user creates objects from both the test_class and example_class classes and invoke the task() method for both of them, the hidden definitions for task() methods inside both the classes come into play. These definitions are hidden from the user. The abstract method task() from the abstract class Absclass is actually never invoked.

But when the print() method is called for both the test_obj and example_obj, the Absclass’s print() method is invoked since it is not an abstract method.

Note: We cannot create instances of an abstract class. It raises an 'Error'.

__Highlight the differences between Encapsulation and Abstraction__

Encapsulation is the packing of "data" and "functions operating on that data" into a single component and restricting the access to some of the object's components.
Encapsulation means that the internal representation of an object is generally hidden from view outside of the object's definition.

Abstraction is a mechanism which represent the essential features without including implementation details.

 - Encapsulation: Information hiding.
 - Abstraction: Implementation hiding.
