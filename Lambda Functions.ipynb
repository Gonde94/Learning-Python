## Lambda Functions (Anonymous Expressions)

A small, one-line anonymous function that is defined without a name. It looks like:

lambda arguments: expression

The function can take multiple arguments, but only one expression.

If you wanted to write a function that calculated 3x + 1, a standard function would:

```python
def standard_function(x):
    return x*3 + 1
```

```python
standard_function(3)
```

```python
#In lambda format

lambda x: 3*x + 1
```

We cannot use the function above because we cannot call it - it has no name. "Lambda" is not the name of the function but a key word in Python. It signifies that what follows is an anonymous function. 

```python
# One way to call it is to give it a name

lambda_function = lambda x: 3*x + 1
lambda_function(3)
```

### Lambda functions with multiple inputs

```python
multi = lambda a, b, c : a + b + c
multi(5, 6, 2)
```

Another example. Combine a person's first and last name into a full name.

```python
full_name = lambda fn, ln: fn.strip().title() + " " + ln.strip().title()
```

```python
full_name("max", " power")
```

### Lambda functions without a variable name

```python
authors = ["Arthur C. Clarke", "William Shakespeare", "Ursula K. Guin", "Robert A. Caro", "Ryszard Kapuscinski",
          "Hermann Hesse", "Carl Gustav Jung", "Laurie Lee", "J.R.R. Tolkien", "Gabriel Garcia Marquez"]
```

We have a list of authors, which we want to alphabetize. Some have middle names or initials, others don't. We can use a lambda function to extract the last name and use that as a sorting value. Call help() on the list.sort and see the arguments needed. We can pass a lambda function in the place of "key". 

```python
help(authors.sort)
```

```python
authors.sort(key = lambda name: name.split(" ")[-1].lower())
```

```python
authors
```

To access the last name, split the name into pieces wherever it has a space (.split(" ")). Access the last piece by index -1. Lastly convert the name to lower case so the function is not case sensitive (.lower()).

### When to use lambda functions?

 - Useful in order to write short, concise code only needed once. 
 - Saves writing multiple lines of code.
 - They're inline functions, therefore they are fast and can reduce execution time.

### Disadvantages

 - They can only have one expression.
 - Readibility issues. It is not always clear what they are doing. Cannot use docstrings to help with documentation, unlike with regular functions.
