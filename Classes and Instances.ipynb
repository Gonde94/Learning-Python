## Object-Oriented Programming

OOP is a programming paradigm (pattern; model) that relies on the concept of classes and objects. It is used to structure a programme into simple, reusable pieces of code blueprints called classes, which are used to create individual instances of objects. 

A class is an abstract blueprint used to create more specific, concrete objects. They often represent broad categories, like 'Car' or 'Dog', that share attributes. These classes define what attributes an instances of this type will have, like 'colour' or 'breed', but not the value of those attributes for a specific object, i.e. 'red' or 'Labrador'. 

Classes can also contain functions, called methods, available only to objects of that type. These functions are defined within the class and perform some action helpful to that specific type of object. For example, our 'Car' class may have a method 'repaint' that changes the colour attribute of our car. This function is helpful only to objects of type Car, so we declare it within the Car class, thus making it a method.

### 1. Classes and Instances

Say we wanted to represent the Greek Gods in python code. We could use a class, because each god will have specific attributes and methods (name, domain, etc.) A class can act as a blueprint when we create each god so that this doesn't have to be done manually from scratch each time. Remember: a class is a blueprint for creating instances. Each new god we create will be an instance of the class 'God'. 

```python
# Create 'God' class
class God:
    
    def __init__(self, name, domain, symbol, animal, roman, tribute):
        self.name = name
        self.domain = domain
        self.symbol = symbol
        self.animal = animal
        self.roman = roman
        self.tribute = tribute
```

The __init__ method is called a 'constructor'. It is a special method called upon an object's creation. It is used in the class to initialise the objects attributes. It serves no other purpose. It is only used within classes. 

When we create methods within a class, they receive the instance as the first argument. By convention this is called 'self'. After self, we can specify what other arguments we want to accept. In this case, we will want the god's name, domain, home, symbol and sacred animal. 

Within the init method, we set all the instances variables. 

```python
# Create instances of the class
god_1 = God('Zeus', 'Sky', 'Lightning Bolt', 'Eagle', 'Jupiter', 100000)
god_2 = God('Apollo', 'Music', 'Lyre', 'Raven', 'Apollo', 40000)
god_3 = God('Athena', 'Wisdom', 'Aegis', 'Owl', 'Minerva', 50000)
```

```python
print(god_1.name)
print(god_2.name)
print(god_3.name)
```

When above is run, god_1 (and 2, 3, etc.) are passed through the class as 'self', and the attributes are set. You can also access the values through:

```python
god_1.__dict__
```

### Methods

All above are attributes of our class. What if we wanted to perform some kind of action? We add methods to our class. For example, below will print out Roman name of the god. Remember to add the instance as the first argument to the method. 

```python
class God:
    
    def __init__(self, name, domain, symbol, animal, roman, tribute):
        self.name = name
        self.domain = domain
        self.symbol = symbol
        self.animal = animal
        self.roman = roman
        self.tribute = tribute
        
    def roman_name(self):
        return self.roman
```

```python
print(god_1.roman_name())
print(god_2.roman_name())
print(god_3.roman_name())
```

### 2. Class Variables

Instance variables are the ones set in the __init__ method, e.g. self.name. They are used for data that is unique to each instance. 

Class variables are those shared among all instances of that class. With an Employee class, it might be a uniform bonus given to all employees. For this example, it will be a percentage indicating the increase in the god's tribute amount. Even though it is a class variable, it is accessed through the instance - self.raise_amount. 

```python
class God:
    
    raise_amount = 1.05
    
    def __init__(self, name, domain, symbol, animal, roman, tribute):
        self.name = name
        self.domain = domain
        self.symbol = symbol
        self.animal = animal
        self.roman = roman
        self.tribute = tribute
        
    def roman_name(self):
        return self.roman
    
    def tribute_raise(self):
        self.tribute = int(self.tribute * self.raise_amount)
```

```python
print(god_1.tribute)
print(god_1.raise_amount)
god_1.tribute_raise()
print(god_1.tribute)
```

This class variable can be updated outside the class like so:

```python
God.raise_amount = 1.07
print(god_1.raise_amount)
```

If I wanted to do this for only one instance:

```python
god_3.raise_amount = 1.10
print(God.raise_amount)
print(god_3.raise_amount)
```

A case where using self. for calling class variable would not make sense: summing up number of gods. There is no case one can imagine where you would want this number different for each god. 

```python
class God:
    
    raise_amount = 1.05
    num_of_gods = 0
    
    def __init__(self, name, domain, symbol, animal, roman, tribute):
        self.name = name
        self.domain = domain
        self.symbol = symbol
        self.animal = animal
        self.roman = roman
        self.tribute = tribute
        
        God.num_of_gods += 1
        
    def roman_name(self):
        return self.roman
    
    def tribute_raise(self):
        self.tribute = int(self.tribute * self.raise_amount)
```

```python
God.num_of_gods
```

### 3. Classmethods and Staticmethods

Regular methods within a class automatically take the instance, 'self', as the first argument. We can change this so that it takes the class as the first argument instead. This is done with classmethods, which is done by adding a decorator, @classmethod. This alters the functionality of the method, so that the method can take in the class. By convention, this argument is called 'cls'. Now, we work with the class instead of the instance.

```python
class God:
    
    raise_amount = 1.05
    num_of_gods = 0
    
    def __init__(self, name, domain, symbol, animal, roman, tribute):
        self.name = name
        self.domain = domain
        self.symbol = symbol
        self.animal = animal
        self.roman = roman
        self.tribute = tribute
        
        God.num_of_gods += 1
        
    def roman_name(self):
        return self.roman
    
    def tribute_raise(self):
        self.tribute = int(self.tribute * self.raise_amount)
        
    @classmethod
    def set_tribute_raise_amount(cls, amount):
        cls.raise_amount = amount
        
# Create instances of the class
god_1 = God('Zeus', 'Sky', 'Lightning Bolt', 'Eagle', 'Jupiter', 100000)
god_2 = God('Apollo', 'Music', 'Lyre', 'Raven', 'Apollo', 40000)
god_3 = God('Athena', 'Wisdom', 'Aegis', 'Owl', 'Minerva', 50000)
```

```python
print(God.raise_amount)
print(god_1.raise_amount)
```

If I wanted to change this raise amount outside the class, I can use the method I just created:

```python
God.set_tribute_raise_amount(1.2)
print(God.raise_amount)
```

So, this will be the same as doing God.raise_amount = 1.2. Another use case of classmethods is to create alternative constructors. For example, if our information was all in a string format for some reason, we could create a classmethod to parse through that. 

```python
class God:
    
    raise_amount = 1.05
    num_of_gods = 0
    
    def __init__(self, name, domain, symbol, animal, roman, tribute):
        self.name = name
        self.domain = domain
        self.symbol = symbol
        self.animal = animal
        self.roman = roman
        self.tribute = tribute
        
        God.num_of_gods += 1
        
    def roman_name(self):
        return self.roman
    
    def tribute_raise(self):
        self.tribute = int(self.tribute * self.raise_amount)
        
    @classmethod
    def set_tribute_raise_amount(cls, amount):
        cls.raise_amount = amount
        
    @classmethod
    def from_string(cls, god_str):
        name, domain, symbol, animal, roman, tribute = god_str.split('-')
        return cls(name, domain, symbol, animal, roman, tribute)
```

```python
god_str_1 = 'Dionysus-Wine-Thyrsus-Panther-Liber-20000'
god_4 = God.from_string(god_str_1)
```

```python
god_4.__dict__
```

So, to parse through the incorrectly formatted string, we created a classmethod 'from_string' to do this. This takes in the string. The function splits that string at the hyphen '-', creates a new god instance, and then returns that instance.

Staticmethods don't pass in any specific argument automatically (as in instance or class - they do take an argument). They behave like regular functions, but are included in the a class because they have some logical connection with the class. 

For example, say we wanted a function that would take in a date and return whether it is a feast day for the gods. That has a logical connection with the God class, but it doesn't depend on any specific instance or class variable. This will need to be a staticmethod. An indication a function should be a staticmethod rather than a classmethod or regular method is if you don't access the instance or the class anywhere within the function.

```python
import datetime

class God:
    
    raise_amount = 1.05
    num_of_gods = 0
    
    def __init__(self, name, domain, symbol, animal, roman, tribute):
        self.name = name
        self.domain = domain
        self.symbol = symbol
        self.animal = animal
        self.roman = roman
        self.tribute = tribute
        
        God.num_of_gods += 1
        
    def roman_name(self):
        return self.roman
    
    def tribute_raise(self):
        self.tribute = int(self.tribute * self.raise_amount)
        
    @classmethod
    def set_tribute_raise_amount(cls, amount):
        cls.raise_amount = amount
        
    @classmethod
    def from_string(cls, god_str):
        name, domain, symbol, animal, roman, tribute = god_str.split('-')
        return cls(name, domain, symbol, animal, roman, tribute)
    
    @staticmethod
    def is_feast_day(day):
        #Say weekdays are workdays and weekends are feast days.
        #In python, saturday and sunday are 5 and 6.
        if day.weekday() == 5 or day.weekday() == 6:
            return True
        else:
            return False
```

```python
test_date = datetime.date(2022, 2, 4)
test_date_2 = datetime.date(2022, 4, 30)

print(God.is_feast_day(test_date))
print(God.is_feast_day(test_date_2))
```
